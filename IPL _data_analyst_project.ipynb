```python
ax=df2['toss_winner'].value_counts().plot.bar(width=0.8)
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x()+0.15, p.get_height()+1))
plt.show()
```

```python
w=df2['toss_winner'].value_counts()
m=pd.concat([df2['team1'],df2['team2']]).value_counts()
ax=(w/m*100).sort_values().plot.bar(color = 'c', width = 0.25)
for p in ax.patches:                       #used to display the values on the top of the bar
    ax.annotate('{:.2f}%'.format(p.get_height()), (p.get_x(), p.get_height()+1),fontsize=9)
plt.show()
```

```python
high_scores=df1.groupby(['match_id', 'inning','batting_team','bowling_team'])['total_runs'].sum().reset_index() 
high_scores.nlargest(10,'total_runs')
```

```python
dismiss=["run out","bowled", "caught", "lbw", "stumped", "caught and bowled", "hit wicket"]
ct=df1[df1["dismissal_kind"].isin(dismiss)]
ax=ct.dismissal_kind.value_counts()[:10].plot.bar(width=0.8,color='black')
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+1))
plt.show()
```

```python
dismissal_kinds = ["bowled", "caught", "lbw", "stumped", "caught and bowled", "hit wicket"] 
ct=df1[df1["dismissal_kind"].isin(dismissal_kinds)]
ax=ct['bowler'].value_counts()[:10].plot.bar(width=0.8,color='pink')
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x()+0.10, p.get_height()+1))
plt.show()
```

```python
ax=df1[df1['batsman_runs']==6].batsman.value_counts()[:10].plot.bar(width=0.8,color='orange')
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x(), p.get_height()+10))
plt.show()
```
